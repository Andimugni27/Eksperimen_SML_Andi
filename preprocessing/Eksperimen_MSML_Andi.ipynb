# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
# Cek informasi dasar dan missing values
df.info()
print("\nJumlah Data Kosong:\n", df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# EDA 1: Distribusi Umur (Univariate)
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribusi Umur Pasien')
plt.show()

# EDA 2: Distribusi Golongan Darah (Kategorikal)
plt.figure(figsize=(8, 4))
sns.countplot(x='Blood Type', data=df)
plt.title('Jumlah Pasien Berdasarkan Golongan Darah')
plt.show()
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
# 1. Menghapus data kosong (jika ada)
df_clean = df.dropna()

# 2. Menghapus kolom yang tidak relevan untuk diprediksi
kolom_dihapus = ['Name', 'Date of Admission', 'Discharge Date', 'Doctor', 'Hospital', 'Room Number']
df_clean = df_clean.drop(columns=[k for k in kolom_dihapus if k in df_clean.columns], errors='ignore')

# 3. Menyimpan hasil preprocessing ke file baru
df_clean.to_csv('rekammedis_preprocessing.csv', index=False)
print("Preprocessing selesai! File rekammedis_preprocessing.csv berhasil dibuat.")
```
